```python
%%writefile code.js

// Part A
const data = [5, 12, 3, 20, 11, 9, 30];

for (const i of data) {
  if (i > 10) { console.log(2 * i); }
}
console.log('\n' + data.filter(i => i > 10).map(i => i * 2) + '\n');


// Part B
const numbers = [1, 2, 3, 4, 5, 6];

console.log(numbers.filter(i => i % 2 == 0).map(i => i * 2).reduce((part, i) => part + i, 0) + '\n');

// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log(users.some(i => i.age > 18));
console.log(users.find(i => i.id == 3));
users.forEach(i => console.log(i.name));
console.log('\n')

// Part D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
